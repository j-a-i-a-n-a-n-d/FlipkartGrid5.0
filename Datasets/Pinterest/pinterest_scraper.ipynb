```python
from pinscrape import pinscrape
import pandas as pd
import os
```

```python
def scrape_from_pinterest(search):
    details = pinscrape.scraper.scrape(search, "output", {}, 7, 10)

    if details["isDownloaded"]:
        print("\nDownloading completed !!")
        print(f"\nTotal urls found: {len(details['extracted_urls'])}")
        print(f"\nTotal images downloaded (including duplicate images): {len(details['url_list'])}")
        print(details)

        # Extract base filenames from URLs
        base_filenames = [os.path.basename(url) for url in details['url_list']]

        # Create a DataFrame
        tdf = pd.DataFrame({
            'keyword': [details['keyword']] * len(base_filenames),
            'base_filename': base_filenames
        })
    else:
        print(f"\nNothing to download ({search})")
        return pd.DataFrame(columns=['keyword', 'base_filename'])
    return tdf
```
